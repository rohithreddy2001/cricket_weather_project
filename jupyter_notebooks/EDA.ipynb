```python
df = pd.read_csv("../csv_files/cricket_ml_ready.csv")
df.head()
```

```python
import plotly.express as px

# 1. Create a permanent text category column in your dataframe first
df["innings_label"] = df["is_batting_first"].map({1: "Batting First", 0: "Chasing"})

# 2. Plot using that explicit text column
fig = px.scatter(
    df, 
    x="temperature_c", 
    y="team_total_score", 
    color="innings_label",  # Use our brand new text column here
    color_discrete_map={"Batting First": "#8B0000", "Chasing": "#000080"},  # Dark Red & Navy Blue
    hover_data=["date", "team_total_score", "overs_played", "run_rate"],
    title="Impact of Temperature on Team Score",
    labels={
        "innings_label": "Match Phase", 
        "temperature_c": "Temperature (°C)", 
        "team_total_score": "Team Total Score"
    },
    trendline="ols"
)

# Thickens the OLS trendlines so they stand out clearly
fig.update_traces(selector=dict(type="scatter", mode="lines"), line=dict(width=3))

fig.show()
```

```python
# Temporary column just for the visualization
df['had_prior_rain'] = df['rain_48hr_mm'].apply(lambda x: 'Rainy ( >0mm )' if x > 0 else 'Dry ( 0mm )')

fig = px.box(
    df, 
    x="had_prior_rain", 
    y="team_total_score", 
    color="is_batting_first",
    points="all", # Shows every single match point next to the box plot
    hover_data=["date", "rain_48hr_mm"],
    title="How Rain in the Prior 48 Hours Affects Final Match Totals",
    labels={"had_prior_rain": "Weather Context", "team_total_score": "Total Runs Scored"}
)
fig.show()
```

```python

# Compute the correlation matrix for numeric columns
corr_matrix = df[['temperature_c', 'humidity_pct', 'rain_48hr_mm', 'overs_played', 'team_avg_run_rate', 'run_rate', 'team_total_score']].corr()

fig = px.imshow(
    corr_matrix,
    text_auto=".2f", # Automatically overlays correlation coefficients
    color_continuous_scale='RdBu_r', # Red-to-Blue scale
    title="Feature Correlation Matrix"
)
fig.show()
```
