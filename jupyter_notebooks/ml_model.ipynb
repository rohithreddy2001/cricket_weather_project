```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score

# Import all of our machine learning model paradigms
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# Try to load XGBoost if you have it installed, otherwise skip it cleanly
try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except ImportError:
    HAS_XGB = False

# ==========================================
# STEP 1: LOAD AND CHRONOLOGICALLY SPLIT
# ==========================================
df = pd.read_csv("../csv_files/cricket_ml_ready_updated.csv")

# Sort matches by date (oldest to newest) to prevent data leakage
df = df.sort_values("date").reset_index(drop=True)

# Select features and target based on your exact specifications
features = ['temperature_c', 'humidity_pct', 'rain_48hr_mm', 'overs_played', 'team_avg_run_rate']
target = 'team_total_score'

X = df[features]
y = df[target]

# Create an 80/20 chronological split point
split_idx = int(len(df) * 0.8)

X_train_raw = X.iloc[:split_idx]
X_test_raw = X.iloc[split_idx:]
y_train = y.iloc[:split_idx]
y_test = y.iloc[split_idx:]

# ==========================================
# STEP 2: SCALE THE FEATURES
# ==========================================
# Normalizes data so large columns don't drown out smaller ones
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.transform(X_test_raw)

# ==========================================
# STEP 3: CREATE THE DICTIONARY OF MODELS
# ==========================================
models = {
    "Linear Regression (OLS)": LinearRegression(),
    "Ridge Regression (L2)": Ridge(alpha=1.0),
    "Lasso Regression (L1)": Lasso(alpha=0.1),
    "K-Nearest Neighbors (KNN)": KNeighborsRegressor(n_neighbors=7),
    "Support Vector Regressor (SVR)": SVR(kernel='rbf', C=100),
    "Single Decision Tree": DecisionTreeRegressor(max_depth=5, random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42),
    "Gradient Boosting (Sklearn)": GradientBoostingRegressor(n_estimators=100, max_depth=4, learning_rate=0.05, random_state=42)
}

# Dynamically add XGBoost to the list if available
if HAS_XGB:
    models["XGBoost Regressor"] = XGBRegressor(n_estimators=100, max_depth=4, learning_rate=0.05, random_state=42)

# ==========================================
# STEP 4: TRAIN, PREDICT, AND EVALUATE
# ==========================================
print(f"🚀 Training {len(models)} models on {len(X_train)} rows... Testing on {len(X_test)} rows...\n")

# We will temporarily collect results as simple lists to make sorting easy
model_names = []
mae_scores = []
r2_scores = []

for name, model in models.items():
    # 1. Train the model
    model.fit(X_train, y_train)
    
    # 2. Make predictions on the test set
    predictions = model.predict(X_test)
    
    # 3. Calculate evaluation metrics
    mae = mean_absolute_error(y_test, predictions)
    r2 = r2_score(y_test, predictions) * 100 # Convert to percentage
    
    # 4. Save calculations to our lists
    model_names.append(name)
    mae_scores.append(mae)
    r2_scores.append(r2)

# ==========================================
# STEP 5: CREATE AND PRINT THE LEADERBOARD
# ==========================================
# Create a clean summary table from our collected lists
leaderboard = pd.DataFrame({
    "Model Name": model_names,
    "Average Error (MAE)": mae_scores,
    "Accuracy (R² Score)": r2_scores
})

# Sort the leaderboard so the most accurate model is at the very top
leaderboard = leaderboard.sort_values(by="Accuracy (R² Score)", ascending=False)

# Formatting output columns cleanly
leaderboard["Average Error (MAE)"] = leaderboard["Average Error (MAE)"].map(lambda val: f"±{val:.1f} runs")
leaderboard["Accuracy (R² Score)"] = leaderboard["Accuracy (R² Score)"].map(lambda val: f"{val:.2f}%")

print("=========================== THE FINAL MASTER BENCHMARK ===========================")
print(leaderboard.to_string(index=False))
print("==================================================================================")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV, PredefinedSplit
from sklearn.metrics import mean_absolute_error, r2_score


df = pd.read_csv("../csv_files/cricket_ml_ready_updated.csv")

# Sort matches chronologically to prevent timeline cheating
df = df.sort_values("date").reset_index(drop=True)

# Define our split index (80% Train, 20% Test)
split_idx = int(len(df) * 0.8)


# Create a temporary view of the training data
df_train_temp = df.iloc[:split_idx]

# Calculate the average team total score for each stadium using only training matches
stadium_averages = df_train_temp.groupby('venue')['team_total_score'].mean()

# Map the averages back to the main DataFrame
df['venue_avg_score'] = df['venue'].map(stadium_averages)

# Fill any missing venues with the overall average score of training matches
overall_train_mean = df_train_temp['team_total_score'].mean()
df['venue_avg_score'] = df['venue_avg_score'].fillna(overall_train_mean)


features = [
    'temperature_c', 
    'humidity_pct', 
    'rain_48hr_mm', 
    'overs_played', 
    'team_avg_run_rate', 
    'venue_avg_score'
]
target = 'team_total_score'

X = df[features]
y = df[target]

# Chronological Train-Test Split
X_train_raw = X.iloc[:split_idx].reset_index(drop=True)
X_test_raw = X.iloc[split_idx:].reset_index(drop=True)
y_train = y.iloc[:split_idx].reset_index(drop=True)
y_test = y.iloc[split_idx:].reset_index(drop=True)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.transform(X_test_raw)


# To tune parameters without using the test set, we split our training set 
# further into training (first 80% of train) and validation (last 20% of train)
val_split_idx = int(len(X_train) * 0.8)

# -1 means the row is used for training, 0 means it is used for validation
test_fold = np.zeros(len(X_train))
test_fold[:val_split_idx] = -1  # Training fold
test_fold[val_split_idx:] = 0   # Validation fold

# Pass this predefined chronological split to GridSearchCV
ps = PredefinedSplit(test_fold)


param_grid = {
    'hidden_layer_sizes': [(128, 64), (64, 32), (100, 50), (128,)],
    'activation': ['relu', 'tanh'],
    'learning_rate_init': [0.001, 0.005, 0.01],
    'solver': ['adam'],
    'max_iter': [500],
    'early_stopping': [True],
    'n_iter_no_change': [25],
    'random_state': [40]
}

# Initialize the base Neural Network model
base_nn = MLPRegressor()


print("🚀 Initiating Chronological Grid Search parameter search...")
grid_search = GridSearchCV(
    estimator=base_nn,
    param_grid=param_grid,
    cv=ps,
    scoring='neg_mean_absolute_error', # Optimize to find the lowest MAE
    n_jobs=-1, # Use all available processors
    verbose=1
)

# Fit the grid search on our training data
grid_search.fit(X_train, y_train)

# Get the best trained model
best_nn_model = grid_search.best_estimator_


# Make predictions on the final chronological test set
nn_predictions = best_nn_model.predict(X_test)

# Calculate final evaluation metrics
nn_mae = mean_absolute_error(y_test, nn_predictions)
nn_r2 = r2_score(y_test, nn_predictions) * 100

print("\n=================================================================")
print("TUNED NEURAL NETWORK PERFORMANCE (WITH STADIUM BASELINE)")
print("=================================================================")
print(f"Optimal Parameters: {grid_search.best_params_}")
print(f"Tuned Neural MAE:   ±{nn_mae:.1f} runs")
print(f"Tuned Neural R²:    {nn_r2:.2f}%")
print("=================================================================")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, PredefinedSplit
from sklearn.metrics import mean_absolute_error, r2_score


df = pd.read_csv("../csv_files/cricket_ml_ready_updated.csv")

# Sort matches chronologically by date to prevent timeline cheating
df = df.sort_values("date").reset_index(drop=True)

# Define our split index (80% Train, 20% Test)
split_idx = int(len(df) * 0.8)


df_train_temp = df.iloc[:split_idx]
stadium_averages = df_train_temp.groupby('venue')['team_total_score'].mean()
df['venue_avg_score'] = df['venue'].map(stadium_averages)

overall_train_mean = df_train_temp['team_total_score'].mean()
df['venue_avg_score'] = df['venue_avg_score'].fillna(overall_train_mean)


features = [
    'temperature_c', 
    'humidity_pct', 
    'rain_48hr_mm', 
    'overs_played', 
    'team_avg_run_rate', 
    'venue_avg_score'
]
target = 'team_total_score'

X = df[features]
y = df[target]

# Split our features and target chronologically
X_train_raw = X.iloc[:split_idx].reset_index(drop=True)
X_test_raw = X.iloc[split_idx:].reset_index(drop=True)
y_train = y.iloc[:split_idx].reset_index(drop=True)
y_test = y.iloc[split_idx:].reset_index(drop=True)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test = scaler.transform(X_test_raw)


# We split the training data further (80% train, 20% validation) chronologically
val_split_idx = int(len(X_train) * 0.8)

# -1 for training rows, 0 for validation rows
test_fold = np.zeros(len(X_train))
test_fold[:val_split_idx] = -1  
test_fold[val_split_idx:] = 0   

ps = PredefinedSplit(test_fold)

param_grid = {
    'n_estimators': [50, 100, 150, 200],         # Number of trees in the forest
    'max_depth': [5, 7, 9, 12, None],            # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],             # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4],               # Minimum samples required at a leaf node
    'random_state': [42]                         # Ensures your results are reproducible
}

# Initialize the base Random Forest
base_rf = RandomForestRegressor()


print("🚀 Initiating Random Forest Grid Search...")
grid_search = GridSearchCV(
    estimator=base_rf,
    param_grid=param_grid,
    cv=ps,
    scoring='neg_mean_absolute_error',          # Looks for the combination with the lowest MAE
    n_jobs=-1,                                   # Uses all available CPU cores for speed
    verbose=1
)

# Fit the grid search on the training data
grid_search.fit(X_train, y_train)

# Extract the best combination model
best_rf_model = grid_search.best_estimator_


predictions = best_rf_model.predict(X_test)

# Calculate final updated metrics
mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions) * 100

print("\n=================================================================")
print("TUNED RANDOM FOREST PERFORMANCE (WITH STADIUM BASELINE)")
print("=================================================================")
print(f"Best Parameters Found: {grid_search.best_params_}")
print(f"Tuned Average Error (MAE):  ±{mae:.1f} runs")
print(f"Tuned Accuracy (R² Score):  {r2:.2f}%")
print("=================================================================")
```
